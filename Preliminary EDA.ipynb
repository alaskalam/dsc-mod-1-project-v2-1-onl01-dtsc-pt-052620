```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

```python
movie_dates_df = pd.read_csv('movie_release_dates.csv', index_col=0)
theaters_df = pd.read_csv('movie_theater_data.csv', index_col=0)
awards_df = pd.read_csv('movie_awards.csv', index_col=0)
actors_df = pd.read_csv('Actors_Table.csv')
directors_df = pd.read_csv('Directors_Table.csv')
imdb_base_df = pd.read_csv('IMDb_base.csv')
imdb_budgets_df = pd.read_csv('IMDb_budgets.csv')
```

```python
#Let's look at the most common release months and release days
movie_months = movie_dates_df['release_month'].value_counts()
plt.figure(figsize=(12,8))
sns.barplot(movie_months.index, movie_months.values)
plt.title("Number of Movies Released by Month")
plt.xlabel("Months of the Year")
plt.ylabel("Numbers of Movies Released")
```

```python
movie_days = movie_dates_df['release_day'].value_counts()
plt.figure(figsize=(12,8))
sns.barplot(movie_days.index, movie_days.values)
plt.title('Number of Movies Released by Day')
plt.xlabel('Days of the Week')
plt.ylabel('Number of Movies Released')
```

```python
#Let'
imdb_budgets_df['Profit'] = imdb_budgets_df['Worldwide Gross'] - imdb_budgets_df['Production Budget']

imdb_budgets_df['Profit_Margin'] = (imdb_budgets_df['Worldwide Gross'] - 
                                    imdb_budgets_df['Production Budget'])/imdb_budgets_df['Worldwide Gross']
```

```python
imdb_budgets_df.set_index('Movie', inplace=True)
awards_df.set_index('film_name', inplace=True)

budgets_and_awards = imdb_budgets_df.join(awards_df, how='inner')
budgets_and_awards.head()
```
